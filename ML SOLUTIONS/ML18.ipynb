```python
x = df.iloc[:,:8]
```

```python
x
```

```python
y=df['Outcome']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x['Glucose'],y,test_size=0.25)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
nb = GaussianNB()
```

```python
# nb.fit(x_train.to_numpy(), y_train)
nb.fit(x_train.to_numpy().reshape(-1, 1), y_train)
```

```python
nb.predict(x_test.to_numpy().reshape(-1, 1))
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(nb.predict(x_test.to_numpy().reshape(-1, 1)),y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(x_train.to_numpy().reshape(-1,1),y_train)
```

```python
p = rf.predict(x_test.to_numpy().reshape(-1,1))
```

```python
accuracy_score(p,y_test)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_predictions(p,y_test)
```
