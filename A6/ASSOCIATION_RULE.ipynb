```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori , association_rules
```

```python
import csv
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("hemanthkumar05/market-basket-optimization")

print("Path to dataset files:", path)
```

```python
dataset = []
with open ('/content/Market_Basket_Optimisation.csv') as file:
  reader = csv.reader(file,delimiter=',')
  for row in reader:
    dataset += [row]
```

```python
dataset
```

```python
te = TransactionEncoder()
```

```python
te.fit(dataset)
```

```python
te.columns_
```

```python
te.columns_
```

```python
x= te.fit_transform(dataset)
df = pd.DataFrame(x,columns=te.columns_)
```

```python
df
```

```python
#RULES OF MAKING APRORI MODEL

freq = apriori(df,min_support=0.01,use_colnames=True)
```

```python
freq
```

```python
#FIND THE RULES
rules = association_rules(freq,metric='confidence',min_threshold=0.25)
```

```python
rules
```
